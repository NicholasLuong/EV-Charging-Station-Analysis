```python
# imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
# Get data
Stations = pd.read_csv('alt_fuel_stations.csv')
pd.set_option('display.max_columns', None)
Stations.head()
```

```python
# Rename columns to snake case
col_list = Stations.columns
new_col_list = []
for col in col_list:
    cleaned_col_name = col.replace(' ', '_')
    new_col_list.append(cleaned_col_name.lower())

print(new_col_list)
Stations.columns = new_col_list
Stations.head()
```

```python
Stations.describe
```

```python
for col in Stations.columns:
    print('Column name:',col)
    print(Stations[col].value_counts(dropna=False).head(5))
    print('-'*50)
    print('')
```

```python
# Drop irrelevant columns
drop_cols = ['intersection_directions', 'plus4', 'access_days_time', 'cards_accepted', 'bd_blends', 'ng_fill_type_code',
             'ng_psi', 'ev_network_web', 'federal_agency_id', 'hydrogen_status_link', 'ng_vehicle_class', 'lpg_primary',
             'e85_blender_pump', 'intersection_directions_(french)', 'access_days_time_(french)', 'bd_blends_(french)',
             'groups_with_access_code_(french)', 'hydrogen_is_retail', 'cng_dispenser_num', 'cng_on-site_renewable_source',
             'cng_total_compression_capacity', 'cng_storage_capacity', 'lng_on-site_renewable_source', 'e85_other_ethanol_blends',
             'ev_pricing_(french)', 'lpg_nozzle_types', 'hydrogen_pressures', 'hydrogen_standards', 'cng_fill_type_code',
              'cng_psi', 'cng_vehicle_class', 'lng_vehicle_class', 'rd_blends', 'rd_blends_(french)', 'rd_blended_with_biodiesel',
              'rd_maximum_biodiesel_level', 'nps_unit_name']

print('Shape of data before drop:', Stations.shape)
Stations = Stations.drop(drop_cols, axis=1)
print('Shape of data after drop:', Stations.shape)
```

```python
# convert open_date to datetime
Stations['open_date'] = pd.to_datetime(Stations['open_date'])

# Create a new column to store the year of the Open_Date
Stations['open_year'] = Stations['open_date'].dt.year

EV_Stations = Stations[Stations['fuel_type_code'] == 'ELEC']

EV_Stations.head()
```

```python
EV_Stations.groupby('ev_network').agg(sum)
```

```python
EV_Station_Counts = EV_Stations.groupby('ev_network').agg(sum)

EV_Station_Counts_DC = EV_Station_Counts[EV_Station_Counts['ev_dc_fast_count'] >=100]
EV_Station_Counts_DC = EV_Station_Counts_DC.sort_values(by = 'ev_dc_fast_count',ascending=False)
EV_Station_Counts_DC
```

```python
g5 = sns.barplot(x=EV_Station_Counts_DC.index, y='ev_dc_fast_count', data=EV_Station_Counts_DC)
plt.xticks(rotation = 75)
plt.title('DC Fast Charger Count by Operators \n(>=100 Chargers)')
plt.ylabel('Count of EV DC Fast Chargers')
plt.xlabel('EV Network')
g5.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.show()
```

```python
EV_Station_Counts = EV_Stations.groupby('ev_network').agg(sum)

EV_Station_Counts_AC = EV_Station_Counts[EV_Station_Counts['ev_level2_evse_num'] >=1000]
EV_Station_Counts_AC = EV_Station_Counts_AC.sort_values(by = 'ev_level2_evse_num',ascending=False)
EV_Station_Counts_AC
```

```python
g0 = sns.barplot(x=EV_Station_Counts_AC.index, y='ev_level2_evse_num', data=EV_Station_Counts_AC)
plt.xticks(rotation = 75)
plt.title('Level 2 AC Count by Operator \n (>=1000 Chargers)')
plt.ylabel('Count of Level 2 AC Chargers')
plt.xlabel('EV Network')
g0.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.show()

### change color pallete to match brand color 
```

```python
EV_Stations_subset = EV_Stations[['ev_dc_fast_count', 'ev_level2_evse_num', 'open_year','state','facility_type','ev_network']]

# group by open_year
grouped_EV_Stations_subset = EV_Stations_subset.groupby('open_year').sum()

# select years after 2007 as cumulative chargers installed up to that point were immaterial (<1000)
grouped_EV_Stations_subset = grouped_EV_Stations_subset[grouped_EV_Stations_subset.index > 2007]

# format years as int
grouped_EV_Stations_subset.index = grouped_EV_Stations_subset.index.astype(int)

grouped_EV_Stations_subset
```

Note - DC charger installations continue to increase annually, while level 2 (home / work) chargers opened were lower in 2022 than 2021. 

This suggests rather than supporting destination charging, companies are trying to implement fast-charging infrastructure. 

```python
g1 = grouped_EV_Stations_subset.plot(kind='bar', stacked=False)
plt.xlabel('Year Opened')
plt.ylabel('Count of DC Fast and Level 2 Chargers')
plt.title('Count of DC Fast and Level 2 Chargers by Year Opened')
g1.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.show()
```
